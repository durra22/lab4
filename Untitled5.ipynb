```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['fare_amount'])

plt.title('Boxplt of Amount')
plt.show()
```

```python
#Points outside the whiskers represent potential outliers.
#These extreme revenue values may influence model predictions.
```

```python
Q1 = df['fare_amount'].quantile(0.25)
Q3 = df['fare_amount'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['fare_amount'] < lower) | (df['fare_amount'] > upper)]
outliers.head(15)
```

```python
#The output displays records considered extreme based on statistical boundaries.
#These may be valid high-value transactions or potential data errors.
```

```python
df_no_outliers = df[(df['fare_amount'] >= lower) & (df['fare_amount'] <= upper)]

print('df with outliers: ',df.shape)
print('df without outliers: ',df_no_outliers.shape)
```

```python
#The dataset size is slightly reduced.
#Removing outliers reduces distortion but may also remove important rare events.

#### Important Note on Removing Outliers
#Not all outliers are errors.
```

```python
lower_cap = df['fare_amount'].quantile(0.05) # 5%
upper_cap = df['fare_amount'].quantile(0.95) # 95%

df_capped = df.copy()
df_capped['fare_amount'] = df_capped['fare_amount'].clip(lower_cap, upper_cap)
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df_capped['fare_amount'])

plt.title('Boxplt of fare_amount')
plt.show()
```

```python
#Outliers in the "amount" column were detected using the IQR method. Values outside the lower and upper bounds were removed to reduce their impact. After removal, the data became more consistent and suitable for analysis.
```

```python
df[['fare_amount','passenger_count']].head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = df[['fare_amount', 'passenger_count']].copy()

df_scaled[['fare_amount', 'passenger_count']] = scaler.fit_transform(df_scaled)

df_scaled.head()
```

```python
# Min-Max scaling is applied to normalize the values of 'fare_amount' and 'passenger_count'
# so that they fall within a range between 0 and 1.
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standardized = df[['fare_amount', 'passenger_count']].copy()

df_standardized[['fare_amount', 'passenger_count']] = scaler.fit_transform(df_standardized)

df_standardized.head()
```

```python
# Standardization is applied to the 'fare_amount' and 'passenger_count' columns
# to transform the data so that it has a mean of 0 and a standard deviation of 1.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.heatmap(df_standardized[['fare_amount','passenger_count']].corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# This heatmap shows the correlation between 'fare_amount' and 'passenger_count'
# to understand the relationship between these variables.
```

```python
from sklearn.decomposition import PCA

X = df_standardized[['fare_amount', 'passenger_count']]

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)

print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.scatter(principal_components[:, 0], principal_components[:, 1])
plt.title("PCA Projection")
plt.xlabel("fare_amount")
plt.ylabel("passenger_count")
plt.show()
```

```python
#The scatter plot of the two principal components shows the projection of the standardized features (fare_amount and passenger_count) onto a reduced two-dimensional space. Each point represents a data instance after transformation using PCA.
 #The distribution of the points appears somewhat dispersed, indicating that the features do not have a strong linear relationship. Since the correlation between the selected features is relatively weak, the principal components do not show clear clustering or strong separation.
 #This suggests that PCA is not significantly compressing the information but still provides a transformed representation of the data.
```

```python
#Conclusion

#In this assignment, several data preprocessing steps were applied to improve the dataset quality.
#Missing values were handled, and feature scaling (standardization) was performed to ensure all variables were on the same scale.
#Principal Component Analysis (PCA) was then applied to reduce the dimensionality of the dataset into two components.
#However, due to the weak correlation between the selected features (fare_amount and passenger_count), PCA did not result in significant dimensionality reduction.
#Despite this, PCA still helped in transforming the data into a new feature space and provided insight into the variance structure of the dataset.
```
